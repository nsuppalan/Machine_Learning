## Pandas and Plot

Pandas provide an easy in-built functionality to draw `matplotlib` plots. We will explore in the below examples.

### Data Loading
Let's load some data

```python
timeseries_df = pd.read_csv('data/daily_min_temperatures.csv')
timeseries_df.head()
```

```python
timeseries_df.info()
```

```python
timeseries_df['Date'] = pd.to_datetime(timeseries_df['Date'], format='%Y-%m-%d')
timeseries_df.set_index('Date', inplace=True)
timeseries_df['DailyMinTemp'] = \
    timeseries_df['DailyMinTemp'].astype(str).astype(float, errors='ignore')

timeseries_df.head()
```

```python
timeseries_df.describe()
```

### Time Series Plot

This sample shows how we plot a time series plot directly from `DataFrame`. Simplest way is to call a plot on the `DataFrame`.

```python
timeseries_df.plot()
```

```python
timeseries_df.plot(figsize=(10, 6), title='Daily minimum temperatures in Melbourne, Australia, 1981-1990', grid=True)
```
